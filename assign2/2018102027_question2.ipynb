### Comparing Optimal Bayes and Naive Bayes using simulated Gaussian data for non-zero covariance.

#### Imports

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.colors as colors
import seaborn as sns
import itertools
from scipy.stats import norm
import scipy.stats
from sklearn.naive_bayes import GaussianNB
from scipy.stats import multivariate_normal

%matplotlib inline
sns.set()
```

#### Data Generation

```python
#data of Gaussian Distributions
np.random.seed(seed = 3)
#distribution1
mu1 = [2, 2]
sigma1 = [[5, -4], [-4, 5]]
x1, y1 = np.random.multivariate_normal(mu1, sigma1, 100).T
#distribution2
mu2 = [-2, -2]
sigma2 = [[3.5, 3], [3, 3.5]]
x2, y2 = np.random.multivariate_normal(mu2, sigma2, 100).T
```

#### Excercise: A) Plot the decision boundary

```python
#%%%%%%%%%%%%%% Excercise: Write the code below %%%%%%%%%%%%%%%%%%%%%%%%%
"""
Code for comparing the decision boundary for the optimal and the naive Bayes classifiers goes here
"""

f, axarr = plt.subplots(1,1)
f.set_figheight(15)
f.set_figwidth(15)

# Naive Bayes

# Reconstructing mean and covariance of class 1 for Naive bayes
x1 = x1.reshape(1,100)
y1 = y1.reshape(1,100)
final = np.vstack([x1,y1])
cov_class1 = np.cov(final)
mean_class1 = final.mean(1)


# Reconstructing mean and covariance of class 2 for Naive bayes
x2 = x2.reshape(1,100)
y2 = y2.reshape(1,100)
final = np.vstack([x2,y2])
cov_class2 = np.cov(final)
mean_class2 = final.mean(1)

                         
# Setting the non-diagonal elements of the covariance matrix as 0 as Naive bayes assumes independance between features
cov_class1[0,1]=0
cov_class1[1,0]=0
cov_class2[0,1]=0
cov_class2[1,0]=0
                         
                         
# Plotting the points and its classes
X = np.linspace(-10, 10, 100)
Y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(X, Y)

# fig = plt.figure(figsize=(10,10)) 

Z = []

# Assigining classes to all the points in the grid and getting the naive bayes boundary                
for x,y in zip(np.ravel(X),np.ravel(Y)):
    
    class1 = multivariate_normal.pdf(np.asarray((x,y)), mean =mean_class1, cov = cov_class1)
    class2 = multivariate_normal.pdf(np.asarray((x,y)), mean = mean_class2, cov = cov_class2)
    if class1>class2:
        Z.append(0)
    else:
        Z.append(1)

Z = np.asarray(Z).reshape(X.shape)

axarr.contourf( X, Y, Z, 2, alpha = .1, colors = ('blue','green','red'))
naive_bayes=axarr.contour( X, Y, Z, 1, alpha = 1, colors = ('blue'))
axarr.clabel(naive_bayes, naive_bayes.levels[::2], inline=True, fmt="Naive Bayes", fontsize=15)
axarr.scatter(x1,y1,label="Training Data Class - 1")
axarr.scatter(x2,y2,label="Training Data Class - 2")
axarr.set_xlabel('x-axis')
axarr.set_ylabel('y-axis')
axarr.set_title('Decision Boundary for Naive Bayes(Blue) v/s Decision Boundary for Optimal Bayes(Magenta)')
plt.legend()
##################################################################################################################

#Optimal Bayes

Z=[]
for x,y in zip(np.ravel(X),np.ravel(Y)):
    
    class1 = multivariate_normal.pdf(np.asarray((x,y)), mean =mu1, cov = sigma1)
    class2 = multivariate_normal.pdf(np.asarray((x,y)), mean = mu2, cov = sigma2)
    if class1>class2:
        Z.append(0)
    else:
        Z.append(1)
Z = np.asarray(Z).reshape(X.shape)

axarr.contourf( X, Y, Z, 2, alpha = .1, colors =  ('yellow'))
opt_bayes=axarr.contour( X, Y, Z, 1, alpha = 1, colors =  ('magenta'))
axarr.clabel(opt_bayes, opt_bayes.levels[::2], inline=True, fmt="Optimal Bayes", fontsize=15)
plt.show()
```

#### Excercise: B) Accuracies

```python
#%%%%%%%%%%%%%% EXCERCISE - Write the code below %%%%%%%%%%%%%%%%%%%%%%%%%%%
"""
Code for comparing the accuracies of the optimal and the naive Bayes classifiers goes here
"""
#Optimal Bayes accuracy


count = 0
Z=[]
for x,y in zip(np.ravel(x1),np.ravel(y1)):
    
    class1 = multivariate_normal.pdf(np.asarray((x,y)), mean =mu1, cov = sigma1)
    class2 = multivariate_normal.pdf(np.asarray((x,y)), mean = mu2, cov = sigma2)
    if class1>class2:
        Z.append(0)
    else:
        count+=1
        Z.append(1)
        
for x,y in zip(np.ravel(x2),np.ravel(y2)):
    
    class1 = multivariate_normal.pdf(np.asarray((x,y)), mean =mu1, cov = sigma1)
    class2 = multivariate_normal.pdf(np.asarray((x,y)), mean = mu2, cov = sigma2)
    if class1>class2:
        count+=1
        Z.append(0)
    else:
        Z.append(1)
print("Optimal Bayes Classifier Accuracy")
print(1-(count/200))


#Naive Bayes accuracy


count = 0
Z=[]
for x,y in zip(np.ravel(x1),np.ravel(y1)):
    class1 = multivariate_normal.pdf(np.asarray((x,y)), mean =mean_class1, cov = cov_class1)
    class2 = multivariate_normal.pdf(np.asarray((x,y)), mean = mean_class2, cov = cov_class2)
    if class1>class2:
        Z.append(0)
    else:
        count+=1
        Z.append(1)
        
for x,y in zip(np.ravel(x2),np.ravel(y2)):
    
    class1 = multivariate_normal.pdf(np.asarray((x,y)), mean =mean_class1, cov = cov_class1)
    class2 = multivariate_normal.pdf(np.asarray((x,y)), mean = mean_class2, cov = cov_class2)
    if class1>class2:
        count+=1
        Z.append(0)
    else:
        Z.append(1)

print("Naive Bayes Classifer Accuracy")
print(1-(count/200))
```
